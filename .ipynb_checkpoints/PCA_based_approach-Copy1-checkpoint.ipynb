```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import scipy
from scipy.optimize import minimize
from scipy import stats
```

```python
main_df = pd.read_csv('main_df_220208.csv')
main_df
```

```python
# try removing the samples with ratio == 0
# only include standard curve samples

temp_df = main_df.loc[(main_df['ratio'] > 0) &
                      (main_df['cell_line'] < 28)]

x = np.log(temp_df.loc[:,['mean_intensity_647','mean_intensity_546']].values * 
           temp_df.loc[:,['area']].values)

X = StandardScaler().fit_transform(x)

pca = PCA()
principalComponents = pca.fit_transform(X)


# plot the results as a scatter plot
plt.scatter(principalComponents[:,0],
            principalComponents[:,1],
            s=5,
            c=temp_df['ratio'])
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.colorbar(label='fraction of edited barcodes')
# plt.savefig('signalrecording_e00_plot2.png', dpi=300, format='png')
plt.show()
print(pca.explained_variance_ratio_)
```

Looks fine to me.

```python
# A slightly more sophisticated scoring (all cell lines)


# use PCA to transform the data
x = np.log(main_df.loc[:,['mean_intensity_647','mean_intensity_546']].values * 
           main_df.loc[:,['area']].values)

X = StandardScaler().fit_transform(x)

pca = PCA()
principalComponents = pca.fit_transform(X)


theta = 0.075 # represents chance of transcription termination
b_t = 10 # represents edited signal from transcribed edited bc
b_o = 2 # represents edited signal from transcribed unedited bc

z = np.random.geometric(p=theta, size=10000) - 1

score_list=[]
for i in range(len(main_df)):
    temp=[]
    for j in range(12):
        if main_df.at[i,'barcode'][j] == '1':
            temp.append(b_t*(z >= (12-j)).sum() / 10000.)
        elif main_df.at[i,'barcode'][j] == '0':
            temp.append(b_o*(z >= (12-j)).sum() / 10000.)
    score = np.sum(temp)
    score_list.append(score)
    
plt.scatter(score_list,principalComponents[:,1], c=main_df['ratio'])
plt.xlabel('Score')
plt.ylabel('PC2')
plt.show()
```

```python
# A slightly more sophisticated scoring (shotgun samples only)
# PCs calculated from the whole data

plt.scatter(np.array(score_list)[main_df.loc[main_df['cell_line'] > 27].index.values],
            principalComponents[main_df.loc[main_df['cell_line'] > 27].index.values,1],
            c=main_df.loc[main_df['cell_line'] >27]['ratio'])
plt.xlabel('Score')
plt.ylabel('PC2')
plt.show()
```

```python
# A slightly more sophisticated scoring (shotgun samples only)
# PCs calculated only from the shotgun data

temp_df = main_df.loc[main_df['cell_line'] > 27]

# use PCA to transform the data
x = np.log(temp_df.loc[:,['mean_intensity_647','mean_intensity_546']].values * 
           temp_df.loc[:,['area']].values)

X = StandardScaler().fit_transform(x)

pca = PCA()
principalComponents = pca.fit_transform(X)


theta = 0.075 # represents chance of transcription termination
b_t = 10 # represents edited signal from transcribed edited bc
b_o = 2 # represents edited signal from transcribed unedited bc

z = np.random.geometric(p=theta, size=10000) - 1

score_list=[]
for i in temp_df.index:
    temp=[]
    for j in range(12):
        if temp_df.at[i,'barcode'][j] == '1':
            temp.append(b_t*(z >= (12-j)).sum() / 10000.)
        elif temp_df.at[i,'barcode'][j] == '0':
            temp.append(b_o*(z >= (12-j)).sum() / 10000.)
    score = np.sum(temp)
    score_list.append(score)
    
plt.scatter(score_list,principalComponents[:,1], c=temp_df['ratio'])
plt.xlabel('Score')
plt.ylabel('PC2')
plt.show()
```
